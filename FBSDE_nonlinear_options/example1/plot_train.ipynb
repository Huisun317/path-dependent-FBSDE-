```python
import pandas as pd
import matplotlib.pyplot as plt 
```

# Example1 

### Method 1

```python
df1=pd.read_csv('./eg1_trained_data/method1_1.csv')
```

```python
df1=df1[['iter','y_pred','loss_var']]
df1.set_index(['iter'],inplace=True)
```

```python
df1.iloc[0:2499:10][['loss_var']].plot()
```

```python
df1.iloc[0:2499:50][['y_pred']].plot(figsize=(10,6))
plt.axhline(0.5823,linestyle='--',color='red')
plt.legend(['Estimated Y0 value', 'Exact solution 0.5828'],fontsize=14)
plt.grid(linestyle='--')
plt.xlabel('Iterations',fontsize=14)
plt.ylabel('$Y_0$ values',fontsize=14)
plt.title('Estimated value for each Interations, example 1')
plt.savefig('eg1_method1.png')
```

### Method 2 

```python
df2=pd.read_csv('./eg1_trained_data/method2_2.csv')
```

```python
df2=df2[['iter','y_pred','loss_var']]
df2.set_index(['iter'],inplace=True)
df2=df2.iloc[100:];
```

```python
upper=df2.loss_var.mean()+2.0*df2.loss_var.mean()
lower=df2.loss_var.mean()-2.0*df2.loss_var.mean()
```

```python
df2=df2[df2.loss_var<=upper]
df2=df2[df2.loss_var>=lower]
```

```python
df2.y_pred.mean()
```

```python
df2[['y_pred']].plot(figsize=(10,6),marker='o')
plt.axhline(0.5823,linestyle='--',color='red')
plt.axhline(0.578,linestyle='--',color='green')
plt.legend(['Estimated Y0 value for each iterations', 'Exact solution 0.5823','Mean of estimated $Y_0$, 0.578'],fontsize=14)
plt.grid(linestyle='--')
plt.xlabel('Iterations',fontsize=14)
plt.ylabel('$Y_0$ values',fontsize=14)
plt.savefig('eg1_method2.png')
```
